# Project: Mad Libs

## Project Instructions

You will be creating a madlibs story. 

Your task:
1.   Write a story with some blanks (at least 5)
2.   Ask the user to input a word based on its part of speech
3.   Once the user has given you all of the words, print your story with user input.
4. At least one input must be an int or a float

Want to make it better? You could...
*   Write a statement so that the proper "a" or "an" is used based on user input
*   Allow the user to choose between different story options
*   Add input validation so that only a number is given when a number is asked for
*   Use output formatting to make the printed story look nice
* Anything else!
